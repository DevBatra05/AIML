```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
pd.read_csv("supermarket.csv")
```

```python
data=pd.read_csv("supermarket.csv")
data.info()
```

```python
#1. line PLot(Matplotlib)
plt.figure()
plt.plot(data['Quantity'],data['Total'],marker='*')
plt.title("Line Plot")
plt.xlabel("Quantity")
plt.ylabel("Total")
plt.show()
```

```python
#2. Bar Chart (Matplotlib)
plt.figure()
plt.bar(data['Unit price'],data['cogs'])
plt.title("Bar Chart")
plt.xlabel("Unit price")
plt.ylabel("cogs")
plt.show()
```

```python
#3. Scatter Plot(Matplotlib)
plt.figure()
plt.scatter(data['Quantity'],data['Unit price'])
plt.title("Scatter Plot")
plt.xlabel("Qunatity")
plt.ylabel("Unit price")
plt.show()
```

```python
#4. Histogram (Seaborn)
plt.figure()
sns.histplot(data['Unit price'],bins=6)
plt.title("Histogram")
plt.xlabel("Unit price")
plt.ylabel("Frequency")
plt.show()
```

```python
#5. Box Plot(Seaborn)
plt.figure()
sns.boxplot(x='Quantity', y='Tax 5%', data=data)
plt.title("Box Plot")
plt.xlabel("Quantity")
plt.ylabel("Tax 5%")
plt.show()
```

```python
#6. Heatmap (seaborn)
plt.figure(figsize=(10, 6))
numeric_data = data.select_dtypes(include=[np.number]) # Select only numeric columns
corr_matrix = numeric_data.corr() # Compute the correlation matrix
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm', linewidths=0.5) # Create heatmap
plt.title("Heatmap of Correlation Matrix")
plt.show()
```

```python
af=data[['Quantity','Unit price','Tax 5%','cogs','gross income','Rating']]
corr=af.corr()
sns.heatmap(corr,annot=True)
plt.show()
```
